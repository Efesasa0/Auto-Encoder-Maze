```python
#%%
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import random
import numpy as np
from torchsummary import summary

#Define the Convolutional Autoencoder
class ConvAutoencoder(nn.Module):
    def __init__(self):
        super(ConvAutoencoder, self).__init__()
       
        #Encoder
        self.conv1 = nn.Conv2d(1, 4, 3, padding=1)
        self.conv2 = nn.Conv2d(4, 8, 3, padding=1)
        self.conv3 = nn.Conv2d(8, 16, 3, padding=1)
        self.pool = nn.AvgPool2d(2, 2)
       
        #Decoder
        self.t_conv1 = nn.ConvTranspose2d(16, 8, 2, stride=2)
        self.t_conv2 = nn.ConvTranspose2d(8, 4, 2, stride=2)
        self.t_conv3 = nn.ConvTranspose2d(4, 1, 2, stride=2)

    def forward(self, x):
        x = F.gelu(self.conv1(x))
        x = self.pool(x)
        x = F.gelu(self.conv2(x))
        x = self.pool(x)
        x = F.gelu(self.conv3(x))
        x = self.pool(x)
        x = F.gelu(self.t_conv1(x))        
        x = F.gelu(self.t_conv2(x))
        x = F.sigmoid(self.t_conv3(x))
        return x

#Instantiate the model
model = ConvAutoencoder()
n = 32
summary(model, (1, n,n))



def create_path(start, max_length, n):
    path = [start]
    x, y = start
    
    counter = 0
    while (x,y) !=  (n-1,n-1) and len(path) < max_length:
        step = random.randint(1,3)
        counter += 1
        if counter > 1000:
            break
        # which way to go
        if random.randint(0,1) == 0:
            if x + step >= n:
                continue
            x = x + step
        else:
            if y + step >= n:
                continue
            y = y + step
        path.append((x,y))    
    return path
# ayseg 0552 504 97 75 

def path_len(path):
    len = 0
    x,y = 0,0
    for p in path:
        x_new,y_new = p
        len += x_new - x + y_new - y
        x, y = x_new, y_new
    return len + 1

def draw_path(maze,path,v=1):
    for i in range(len(path)-1):
        x,y = path[i]
        xnew,ynew=path[i+1]
        if x == xnew:
            for j in range(y,ynew+1):
                maze[x,j] = v
        else:
            for i in range(x,xnew+1):
                maze[i,y] = v
    return maze 

def scramble(maze):
    n, m = maze.shape
    for i in range(40):
        x, y = random.randint(0,n-1), random.randint(0,n-1), 
        z = random.randint(1,10)
        if random.randint(0,1) == 0:
            for i in range(x,min(n,x+z)):
                maze[i,y] = 1
        else:
            for i in range(y,min(n,y+z)):
                maze[x,i] = 1
    return maze

def create_maze_with_solution(n):
    maze = torch.zeros(n,n)
    solution = torch.zeros(n,n)
    sol_path = create_path((0,0),2*n-1,n)
    draw_path(solution, sol_path,v=1)
    draw_path(maze, sol_path,v=1)
    scramble(maze)
    return (maze, solution)

def create_training_maze(m,n):
    mazes = torch.zeros(m,1,n,n)
    sols = torch.zeros(m,1,n,n)
    for i in range(m):
        m, s = create_maze_with_solution(n)
        mazes[i,0,:,:] = m
        sols[i,0,:,:] = s
    return mazes, sols

# Provide a test "maze with noise" to solve
test_maze = torch.zeros(1,n,n)
test_maze[0,:,:], dummy = create_maze_with_solution(n)

#0212 283 52 22
mazes, sols = create_training_maze(10000,n)

dataset = torch.utils.data.TensorDataset(mazes, sols)
dataloader = torch.utils.data.DataLoader(dataset, batch_size=32, num_workers=0)

#Loss function
criterion = nn.BCELoss()

#Optimizer
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

def test_model():
    test_maze = torch.zeros(1,n,n)
    test_maze[0,:,:], sol = create_maze_with_solution(n)
    test_maze_solved = model(test_maze).detach().numpy()
    #test_maze_solved[test_maze_solved < 0.5] = 0
    #test_maze_solved[test_maze_solved >= 0.5] = 1
    fig, axes = plt.subplots(1,3)
    axes[0].imshow(test_maze[0,:,:])
    axes[1].imshow(test_maze_solved[0,:,:])
    axes[2].imshow(sol)
    
    plt.show()

def get_device():
    if torch.cuda.is_available():
        device = 'cuda:0'
    else:
        device = 'cpu'
    return device

device = get_device()
print(device)
model.to(device)

n_epochs = 100

for epoch in range(1, n_epochs+1):
    # monitor training loss
    train_loss = 0.0

    #Training
    for images, targets in dataloader:
        # images are mazes with noise
        # targets are mazes without noise
        images = images.to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, targets)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()*images.size(0)
          
    train_loss = train_loss/len(dataloader)
    print('Epoch: {} \tTraining Loss: {:.6f}'.format(epoch, train_loss))
    test_model()
```
